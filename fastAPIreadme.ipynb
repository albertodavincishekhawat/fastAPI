- what are <u>path operations</u>?
<br>
    - In the context of APIs, "path operations" typically refer to the HTTP methods (such as GET, POST, PUT, DELETE, etc.) and the corresponding paths (URLs) that are used to perform actions or operations on a resource. These operations define how clients can interact with the API endpoints to retrieve, create, update, or delete data.(also called <b>ROUTES</b>)
    - <u>For Example</u>, in a RESTful API design, you might have path operations like:

        1. GET /users - Retrieve a list of users.
        2. GET /users/{id} - Retrieve details of a specific user.
        3. POST /users - Create a new user.
        4. PUT /users/{id} - Update an existing user.
        5. DELETE /users/{id} - Delete a user.

    - Each path operation corresponds to a specific action that can be performed on a resource (in this case, users), and the HTTP method used indicates the type of operation being requested. These operations form the basic building blocks of API endpoints, allowing clients to interact with the API and manipulate resources as needed.
